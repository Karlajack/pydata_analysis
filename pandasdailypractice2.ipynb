Ex1 - Filtering and Sorting Data

```python
url='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

chipo=pd.read_csv(url,sep='\t')
chipo
```

Step 4. How many products cost more than $10.00?

```python
chipo
```

Step 4. How many products cost more than $10.00?

```python
chipo['item_price'>10]
```

Step 5. What is the price of each item?

```python
chipo[['item_name', 'item_price']]
```

Step 6. Sort by the name of the item

```python
chipo.sort_values('item_name')
```

Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.groupby('item_price').quantity.sum().sort_index(ascending = False).iloc[0]
```

Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie=chipo[(chipo.item_name == 'Veggie Salad Bowl')]

veggie.groupby('item_name').order_id.count()


```

Step 9. How many times did someone order more than one Canned Soda?

```python
cannedsoda=chipo[(chipo.item_name == 'Canned Soda') & (chipo.quantity > 1)]

cannedsoda.groupby('item_name').item_name.count()
```
